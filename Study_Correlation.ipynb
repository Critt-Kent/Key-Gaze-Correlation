## Read and prepare data

```python
# this reads the clustered and HOF annotated data
AUdf = pd.read_csv("../copyfiles/AU_cluster5_PUB", sep="\t") 
print(AUdf.shape)
AUdf.head()
```

- sequences of AUs of type 1 (ST reading), 2 (TT reading), are the ones that cluster into pauses (KBIs or PUBs).

```python
# Filter for pause-like AUs (Types 1, 2, 5, 6) 

pause_au = AUdf[AUdf['Type'].isin([1, 2, 5, 6])].copy()
```

## Calculating average for Dur_L, Dur_S, Dur_R for each session

```python
pause_au_new = pause_au.copy()
dur_cols = ['Dur_L', 'Dur_R', 'Dur_S']
new_avg_cols = ['avg_' + col for col in dur_cols]

if new_avg_cols[0] not in pause_au_new.columns:
    
    # Calculate the average for each column, grouped by 'StudySession'
    avg_df = pause_au_new.groupby('StudySession')[dur_cols].mean().reset_index()

    # Rename the columns to the desired format (e.g., 'Dur_L' -> 'avg_L')
    avg_df.columns = ['StudySession'] + new_avg_cols

    # Merge the new average columns back into the copied DataFrame
    pause_au_new = pd.merge(
        pause_au_new,
        avg_df,
        on='StudySession',
        how='left'
    )
    print("Average columns added successfully.")
else:
    print("Average columns already exist. Skipping merge operation.")

pause_au_new.head()
```

## Calculations per study

###  Descriptive Statistics & Visualization

```python
print("--- Descriptive Statistics & Visualization ---")

# Calculate descriptive statistics for gaze features, grouped by Study
gaze_stats = pause_au_new.groupby('Study')[['avg_Dur_L', 'avg_Dur_R', 'avg_Dur_S']].describe()
print("Summary statistics for gaze durations by pause type:")
gaze_stats
```

### Correlations

```python
corr_final = pause_au_new[['KBI','PUB','avg_Dur_L','avg_Dur_R','avg_Dur_S']]
corr_final.corr()
```

```python
# Get the list of unique Studys
study_to_process = pause_au_new['Study'].unique()

# columns to calculate correlations
columns_to_correlate = ['Dur', 'avg_Dur_L', 'avg_Dur_R', 'avg_Dur_S', 'KBI', 'PUB']

results_list = []
summary_results = []
# Loop through each Study to calculate per-translator correlations
for study_id in study_to_process:
    # Filter the DataFrame for the current study
    session_data = pause_au_new[pause_au['Study'] == study_id]
    
    # Check if there's enough data to compute a correlation
    # A correlation needs at least 2 data points for each column
    if len(session_data) > 1:
        # Calculate the correlation matrix for the selected columns
        correlation_matrix = session_data[columns_to_correlate].corr()

        # Create a dictionary to store the correlations for the current study
        study_correlations = {'Study': study_id}
        
        # Manually extract the specific correlation pairs
        study_correlations['Dur_vs_KBI'] = correlation_matrix.loc['Dur', 'KBI']
        study_correlations['Corr_Dur_L_vs_KBI'] = correlation_matrix.loc['avg_Dur_L', 'KBI']
        study_correlations['Corr_Dur_R_vs_KBI'] = correlation_matrix.loc['avg_Dur_R', 'KBI']
        study_correlations['Corr_Dur_S_vs_KBI'] = correlation_matrix.loc['avg_Dur_S', 'KBI']
        study_correlations['Corr_KBI_vs_PUB'] = correlation_matrix.loc['KBI', 'PUB']
        study_correlations['Corr_Dur_L_vs_PUB'] = correlation_matrix.loc['avg_Dur_L', 'PUB']
        study_correlations['Corr_Dur_R_vs_PUB'] = correlation_matrix.loc['avg_Dur_R', 'PUB']
        study_correlations['Corr_Dur_S_vs_PUB'] = correlation_matrix.loc['avg_Dur_S', 'PUB']
        
        summary_results.append(study_correlations)
        results = {
            'Study': study_id,
            'CorrelationMatrix': correlation_matrix
        }
        results_list.append(results)
    else:
        print(f"Not enough data for study {study_id}. Skipping correlation.")

# Convert the list of dictionaries into a DataFrame
summary_df = pd.DataFrame(summary_results)

for result in results_list:
    print(f"\n--- Correlation Matrix for {result['Study']} ---")
    print(result['CorrelationMatrix'])
```
