```python
# Filter for pause-like AUs (Types 1, 2, 5, 6) 

pause_au = AUdf[AUdf['Type'].isin([1, 2, 5, 6])].copy()
```

```python
# Get the list of unique Studys
study_to_process = pause_au['Study'].unique()

# columns to calculate correlations
columns_to_correlate = ['Dur', 'Dur_L', 'Dur_R', 'Dur_S', 'KBI', 'PUB']

results_list = []
summary_results = []
# Loop through each Study to calculate per-translator correlations
for study_id in study_to_process:
    # Filter the DataFrame for the current study
    session_data = pause_au[pause_au['Study'] == study_id]
    
    # Check if there's enough data to compute a correlation
    # A correlation needs at least 2 data points for each column
    if len(session_data) > 1:
        # Calculate the correlation matrix for the selected columns
        correlation_matrix = session_data[columns_to_correlate].corr()

        # Create a dictionary to store the correlations for the current study
        study_correlations = {'Study': study_id}
        
        # Manually extract the specific correlation pairs
        study_correlations['Dur_vs_KBI'] = correlation_matrix.loc['Dur', 'KBI']
        study_correlations['Corr_Dur_L_vs_KBI'] = correlation_matrix.loc['Dur_L', 'KBI']
        study_correlations['Corr_Dur_R_vs_KBI'] = correlation_matrix.loc['Dur_R', 'KBI']
        study_correlations['Corr_Dur_S_vs_KBI'] = correlation_matrix.loc['Dur_S', 'KBI']
        study_correlations['Corr_KBI_vs_PUB'] = correlation_matrix.loc['KBI', 'PUB']
        study_correlations['Corr_Dur_L_vs_PUB'] = correlation_matrix.loc['Dur_L', 'PUB']
        study_correlations['Corr_Dur_R_vs_PUB'] = correlation_matrix.loc['Dur_R', 'PUB']
        study_correlations['Corr_Dur_S_vs_PUB'] = correlation_matrix.loc['Dur_S', 'PUB']
        
        summary_results.append(study_correlations)
        results = {
            'Study': study_id,
            'CorrelationMatrix': correlation_matrix
        }
        results_list.append(results)
    else:
        print(f"Not enough data for study {study_id}. Skipping correlation.")

# Convert the list of dictionaries into a DataFrame
summary_df = pd.DataFrame(summary_results)

for result in results_list:
    print(f"\n--- Correlation Matrix for {result['Study']} ---")
    print(result['CorrelationMatrix'])
```
